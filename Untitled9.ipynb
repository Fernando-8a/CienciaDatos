# Exploracion de la Estructura del Dataset 
##  ¿Qué hace `df.info()`?

El comando `df.info()` muestra información general del DataFrame, incluyendo:

- Número total de filas.
- Número total de columnas.
- Nombre de cada columna.
- Tipo de dato de cada columna (int, float, object, etc.).
- Cantidad de valores NO nulos por columna.
- Memoria utilizada.

** En resumen:** sirve para conocer la estructura del dataset y detectar datos faltantes o tipos incorrectos.

##  ¿Qué hace `df.describe()`?

El método `df.describe()` en Pandas genera un resumen **estadístico descriptivo** de las columnas del DataFrame.

Por defecto, incluye las columnas numéricas y calcula medidas como:  
- `count` (cantidad de valores no nulos)  
- `mean` (promedio)  
- `std` (desviación estándar)  
- `min` (valor mínimo)  
- `25%`, `50%` y `75%` (percentiles)  
- `max` (valor máximo) :contentReference[oaicite:0]{index=0}  

También tiene parámetros para modificar su comportamiento:  
- `percentiles`: permite definir qué percentiles incluir. :contentReference[oaicite:1]{index=1}  
- `include` / `exclude`: para especificar tipos de columnas que quieres resumir (por ejemplo `'all'` para incluir también columnas categóricas). :contentReference[oaicite:2]{index=2}  

Si incluyes columnas no numéricas (por ejemplo `include='all'`), el resultado mostrará estadísticas como:  
- `unique` (número de valores distintos)  
- `top` (el valor más frecuente)  
- `freq` (frecuencia del valor más frecuente) :contentReference[oaicite:3]{index=3}  

**En resumen:** `df.describe()` te da una visión rápida de la tendencia central, dispersión y rango de tus datos numéricos, y también puede dar resumen para datos categóricos si ajustas los parámetros.  



```python
df= pd.read_csv(archivo)
print("\nPrimeras filas:")
print(df.head())

print("\nInformación de las columnas:")
df.describe()
```

##  ¿Qué hace `df.isnull().sum()`?

- `df.isnull()` crea un DataFrame booleando: cada celda es `True` si el valor original es nulo (NaN o None), o `False` si no lo es. :contentReference[oaicite:0]{index=0}  
- Luego, al aplicar `.sum()`, Pandas suma esos valores booleanos por columna. En Python, `True` equivale a 1 y `False` a 0, por lo que el resultado es el **conteo de valores nulos en cada columna**. :contentReference[oaicite:1]{index=1}  
- El resultado es una **Series de Pandas** donde cada índice es el nombre de una columna y el valor es la cantidad de valores faltantes en esa columna. :contentReference[oaicite:2]{index=2}  

**En resumen:** este comando te ayuda a identificar cuántos valores nulos hay en cada columna de tu DataFrame, lo que es muy útil para la limpieza de datos.


```python
df= pd.read_csv(archivo)
print("\nPrimeras filas:")
print(df.head())

print("\nInformación de las columnas:")
df.isnull().sum()
```

## B) Boxplots

```python

num_vars = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

plt.figure(figsize=(15,10))
for i, var in enumerate(num_vars):
    plt.subplot(4, 3, i+1)  # Organiza los gráficos en una cuadrícula
    sns.boxplot(y=df[var], color='lightgreen')
    plt.title(f"Boxplot de {var}")

plt.tight_layout()
plt.show()
```

¿Hay valores atípicos?  
- `mpg`: Algunos autos muy eficientes aparecen como outliers.  
- `cyl`: No hay valores extremos importantes, la mayoría de autos tiene 4, 6 u 8 cilindros.  
- `disp`: Varios outliers hacia valores altos, motores muy grandes.  
- `hp`: Presenta outliers en autos con potencia muy alta.  
- `drat`: Distribución compacta, sin outliers relevantes.  
- `wt`: Algunos autos muy pesados son outliers.  
- `qsec`: Distribución bastante uniforme, pocos outliers.  
- `vs`, `am`: Variables binarias, no aplican outliers.  
- `gear`: No hay outliers relevantes; valores concentrados en 3 y 4 marchas.  
- `carb`: Algunos autos con muchos carburadores son outliers.

 ¿Cómo se ubica la mediana?  
- En general, la mediana se encuentra dentro del rango intercuartílico (`25% - 75%`).  
- En variables sesgadas como `hp` o `disp`, la mediana está más cerca del límite inferior del boxplot, mostrando que la cola superior eleva la media pero no la mediana.

## C) Gráfico de barras (Categoría)

```python

df['make'] = df['Unnamed: 0'].str.split().str[0]

top_makes = df['make'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_makes.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 marcas de autos más frecuentes", fontsize=14)
plt.xlabel("Marca", fontsize=12)
plt.ylabel("Número de modelos", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## ¿Qué categoría tiene más observaciones y por qué?

| Marca       | Frecuencia | Interpretación |
|---|---|---|
| **Mercedes** | Es la más alta en la tabla de marcas | Mercedes es la marca dominante en el dataset, lo que indica que muchos de los modelos de autos presentes pertenecen a esa marca. En términos estadísticos, Mercedes representa la **moda** de la variable “marca”, es decir, el valor más frecuente en esa variable. |

**Razones por las que Mercedes tiene más observaciones:**

1. **Moda en variable cualitativa:** La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. En este caso, es Mercedes. 
2. **Dominancia en la muestra:** La alta frecuencia sugiere que tu dataset está “sesgado” hacia Mercedes, lo que podría ser porque tienes muchos modelos de esa marca o porque esa marca es más representativa en la población de autos que estás analizando.  
3. **Importancia analítica:** Identificar la marca más frecuente es útil para el análisis exploratorio, ya que puedes investigar si los modelos de Mercedes tienen características particulares (como potencia, consumo, peso) en comparación con otras marcas, dado que su alta representación podría influir en las tendencias generales del dataset.

> En conclusión, la categoría **Mercedes** tiene más observaciones porque es la más frecuente (“moda”) en la variable de marca, lo que la convierte en un punto clave para enfocarte en posteriores análisis exploratorios.

## D) Gráfico de dispersión

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='hp', y='mpg', data=df)
plt.title("Dispersión: Caballos de fuerza (hp) vs Millas por galón (mpg)")
plt.xlabel("Potencia (hp)")
plt.ylabel("Eficiencia (mpg)")
plt.show()
```

## ¿Se observa alguna tendencia o patrón general?

- En el scatterplot de **caballos de fuerza (hp)** frente a **millas por galón (mpg)**, se aprecia claramente una **tendencia negativa**: a medida que la potencia del motor (hp) aumenta, el mpg tiende a disminuir.  
- Esto sugiere una **correlación inversa**: autos con más potencia generalmente tienen menor eficiencia de combustible. Esto tiene sentido porque para generar más potencia el motor consume más combustible. :contentReference[oaicite:0]{index=0}  
- También se puede notar cierta dispersión en los puntos: no todos los autos con alta potencia tienen exactamente el mismo bajo mpg, lo que indica variabilidad entre modelos.
- Además, podrían existir algunos **outliers**: coches con mucha potencia pero no tan bajo mpg, o autos con potencia moderada pero eficiencia inusualmente baja, lo que podría merecer un análisis más detallado.
